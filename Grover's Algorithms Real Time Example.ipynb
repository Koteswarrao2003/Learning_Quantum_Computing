```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
from qiskit import transpile
```

```python
data=[2,4,1,4,6,88,4,2,8,9,0,7,6,44,6,63,88,5,33,88,3,6,6,77,64]
target=77
target_index=data.index(target)

print(target)
print(target_index)
```

```python
binaryString=format(target_index,'05b')
print(binaryString)
```

```python
n=5
circuit=QuantumCircuit(n,n)
circuit.h(range(n))

#oracle State
circuit.x(3)
circuit.h(4)
circuit.mcx([0,1,2,3],4)
circuit.h(4)
circuit.x(3)
circuit.draw(output='mpl')
```

```python
#Diffusion operator
circuit.h(range(n))
circuit.x(range(n))
circuit.h(4)
circuit.mcx([0,1,2,3],4)
circuit.h(4)
circuit.x(range(n))
circuit.h(range(n))
circuit.draw(output='mpl')
```

```python
circuit.measure(range(n),range(n))
circuit.draw(output='mpl')
```

```python
from qiskit_aer import Aer
simulator=Aer.get_backend('qasm_simulator')
circuit=transpile(circuit,simulator)
job=simulator.run(circuit, shots=1024)
result=job.result()
count=result.get_counts()
plot_histogram(count)
```
